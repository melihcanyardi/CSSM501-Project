# Ekşi Web Scraping - Gündem

```python
import pandas as pd
import waybackpy
from bs4 import BeautifulSoup
import time
```

```python
def get_topic_list_from_date(date):
    wayback = waybackpy.Url("https://eksisozluk.com", "Chrome/96.0.4664.45")
    timestamp = pd.to_datetime(wayback.near(year=date.year, month=date.month, day=date.day, hour=12, minute=0).timestamp)
    soup = BeautifulSoup(wayback.near(year=date.year, month=date.month, day=date.day).get(), 'html.parser')
    
    urls = [x.get("href")[20:-10] for x in soup.find("ul", "topic-list partial").find_all("a")]
    topics = [' '.join(x.get_text().split()[:-1]) for x in soup.find("ul", "topic-list partial").find_all("a")]
    entry_counts = [x.get_text().split()[-1] for x in soup.find("ul", "topic-list partial").find_all("a")]
    
    return [timestamp] * len(urls), urls, topics, entry_counts
```

```python
date_range = pd.date_range("2021-01-01", "2022-01-01", freq="D", closed="left")
```

```python
timestamps_list = []
urls_list = []
topics_list = []
entry_counts_list = []
sponsored_timestamps_list = []
```

```python
for i, date in enumerate(date_range):
    try:
        timestamps, urls, topics, entry_counts = get_topic_list_from_date(date)
        timestamps_list.extend(timestamps)
        urls_list.extend(urls)
        topics_list.extend(topics)
        entry_counts_list.extend(entry_counts)
        print(f"{i+1}- {timestamps[0]} | {(i+1)/len(date_range)*100:.2f}%")
    except:
        time.sleep(5)
        timestamps, urls, topics, entry_counts = get_topic_list_from_date(date)
        timestamps_list.extend(timestamps)
        urls_list.extend(urls)
        topics_list.extend(topics)
        entry_counts_list.extend(entry_counts)
        print(f"{i+1}- {timestamps[0]} | {(i+1)/len(date_range)*100:.2f}%")
    else:
        continue
```

```python
# Creating a DataFrame from Topics, URLs, Entry Counts, and Snapshot Times
eksi_df = pd.DataFrame([topics_list, urls_list, entry_counts_list, timestamps_list]).T.rename({0: 'Topics', 1: 'URL', 2: 'Entry_Counts', 3: "Snapshot_Time"}, axis=1)

# Dropping sponsored content days
drop_indices = []

for i, count in enumerate(eksi_df["Entry_Counts"]):
    try:
        int(count)
    except:
        if len(count) == 4 and count[-1] == "b" and count[1] == ",":
            pass
        else:
            drop_indices.append(i)

eksi_df = eksi_df.drop(index=drop_indices).reset_index(drop=True)

## Changing entries with counts like "1,2b" into "1200"
eksi_df["Entry_Counts"] = eksi_df["Entry_Counts"].apply(lambda x: int(''.join(x[:-1].split(',') + ['0', '0', '0'])) if x[-1] == "b" else x).astype(int)

eksi_df = eksi_df.drop_duplicates()
```

```python
eksi_df.to_csv("eksi_gundem.csv", index=False)
```
